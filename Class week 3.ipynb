```python
# challenge 3: variable naming and arithmetic operations
price_apples = 1.5
num_apples = float(12)
total_cost = price_apples * num_apples
print("The total cost for {} apples is ${}.".format(int(num_apples),total_cost))
print(f"The total cost for {num_apples} apples is ${total_cost}")
print(type(price_apples))
print(type(num_apples))
print(type(total_cost))
```

```python
# challenge 4: updating variables
current_age = 45
print("My current age is: {}.".format(current_age))
current_age += 5
print("My new current age is: {}.".format(current_age))
print(type(current_age))
```
